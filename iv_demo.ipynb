```python
import pandas as pd
import numpy as np
import copy
import time
```

```python
def cal_iv(data,target='label',columns=[],min_example=0.05,alpha=0.05):
    import woe.feature_process as fp
    from woe.eval import eval_feature_detail
    """
    param:
        data:数据
        target:分类标签列名
        columns:需要计算iv值的列
        min_example:每个bin中最小样本量，一般设为样本总量的5%
        alpha: 自动分箱标准
    return:
        pd.dataFrame
    """
    if len(columns)==0:
        columns = data.columns
    data['target'] = data[target]
    data_woe = data # is used to store the woe value of all data
    civ_list = []
    n_positive = sum(data[target])
    n_negtive = len(data) - n_positive
    for column in list(columns):
        if data[column].dtypes == 'object':
            civ = fp.proc_woe_discrete(data, column, n_positive, n_negtive, min_example*len(data), alpha=alpha)
        else:            
            civ = fp.proc_woe_continuous(data, column, n_positive, n_negtive, min_example*len(data), alpha=alpha)
        civ_list.append(civ)
        data_woe[column] = fp.woe_trans(data[column], civ)
    civ_df = eval_feature_detail(civ_list)
    iv = civ_df[['var_name','iv']].drop_duplicates()
    return iv
```

```python
df = pd.read_csv('data/data.csv')
now = time.time()
iv = cal_iv(copy.deepcopy(df))
print(time.time() - now)
```
